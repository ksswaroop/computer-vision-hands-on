# Image Segmentation based on Edge based and Region Based

```python
#Import the required packages
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
#Import image from skimage
#from skimage import data
img=cv.imread("C:/Users/ksais/OneDrive/Documents/Coding/Computer Vision/computer-vision-hands-on/Q4/dataset/Image 3.jpg")
hist=np.histogram(img,bins=np.arange(0,256))
fig,(ax1)=plt.subplots()
#ax1.imshow(coins)
ax1.imshow(img,cmap=plt.cm.gray,interpolation='nearest')
```

## Edge-based segmentation

```python
#Importing canny edge detection
from skimage.feature import canny
#Convert Image from RGB to Gray scale
gray_image = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
edges = canny(gray_image/255)
fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(edges, cmap=plt.cm.gray, interpolation='nearest')
ax.axis('off')
ax.set_title('Canny detector')
#Text(0.5, 1.0, 'Canny detector')
```

#### Small spurious objects are easily removed by setting a minimum size for valid objects:

```python
from scipy import ndimage as ndi
fill_img = ndi.binary_fill_holes(edges)
fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(fill_img, cmap=plt.cm.gray, interpolation='nearest')
ax.axis('off')
ax.set_title('Filling the holes')
```

## Region-based segmentation

```python
#Importing the required poackages
from skimage.filters import sobel
img=cv.imread("C:/Users/ksais/OneDrive/Documents/Coding/Computer Vision/computer-vision-hands-on/Q4/dataset/Image 3.jpg")
#Convert RGB image to gray scale
gray_image = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
#Applyiung Sobel filter
elevation_map = sobel(gray_image)
fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(elevation_map, cmap=plt.cm.gray, interpolation='nearest')
ax.axis('off')
ax.set_title('elevation_map')
```
