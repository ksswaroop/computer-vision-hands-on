```python
def spatial_fuzzy_clustering(img, k, m, eps):
    # Convert the image to a floating-point format
    img = np.float32(img)

    # Initialize the membership matrix
    membership = np.zeros((img.shape[0], img.shape[1], k))

    # Initialize the cluster centers randomly
    centers = np.random.rand(k, 2)

    # Loop until convergence
    while True:
        # Compute the distances between each pixel and each cluster center
        dist = np.zeros((img.shape[0], img.shape[1], k))
        for i in range(k):
            dist[:, :, i] = np.sqrt((img - centers[i, 0])**2 + (np.indices(img.shape)[0] - centers[i, 1])**2)

        # Compute the new membership matrix
        new_membership = 1 / np.sum((dist / eps)**(2/(m-1)), axis=2)
        new_membership = new_membership / np.sum(new_membership, axis=2)[:,:,np.newaxis]

        # Compute the new cluster centers
        new_centers = np.zeros((k, 2))
        for i in range(k):
            new_centers[i, 0] = np.sum(new_membership[:, :, i] * img) / np.sum(new_membership[:, :, i])
            new_centers[i, 1] = np.sum(new_membership[:, :, i] * np.indices(img.shape)[0]) / np.sum(new_membership[:, :, i])

        # Check for convergence
        if np.max(np.abs(new_membership - membership)) < eps:
            break

        membership = new_membership
        centers = new_centers

    return membership, centers
```

```python
# Load the input image
#img = cv2.imread('input_image.png', cv2.IMREAD_GRAYSCALE)
# Perform spatial fuzzy clustering
k = 2
m = 2
eps = 0.001
membership, centers = spatial_fuzzy_clustering(img1, k, m, eps)

# Threshold the membership matrix to obtain a binary mask
mask = np.argmax(membership, axis=2)


# Apply morphological operations to clean up the mask
kernel = np.ones((5, 5), np.uint8)
mask = cv2.morphologyEx(mask.astype(np.uint8), cv2.MORPH_OPEN, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

# Save the output mask
cv2.imwrite('output_mask.png', mask*255)
```
